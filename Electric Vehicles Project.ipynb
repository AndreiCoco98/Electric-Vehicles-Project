```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\coco MANEL\Downloads\Electric_Vehicle_Population_Data.csv")
df
```

```python
# Assuming df is your large DataFrame
# Extracting specific columns
columns_to_extract = ['Model', 'Model Year', 'Electric Range']
df2 = df[columns_to_extract]

# Now df2 contains only the specified columns from df
```

```python
df2
```

```python
# Group by year and calculate the average electric range
avg_range_by_year = df2.groupby('Model Year')['Electric Range'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(avg_range_by_year['Model Year'], avg_range_by_year['Electric Range'], marker='o', linestyle='-')
plt.title('Average Electric Range by Year')
plt.xlabel('Year')
plt.ylabel('Average Electric Range (km)')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Group by car model and calculate the average electric range
avg_range_by_model = df.groupby('Model')['Electric Range'].mean().reset_index()

# Sort the DataFrame by electric range in descending order and select the top 10 rows
top_10_avg_range = avg_range_by_model.sort_values(by='Electric Range', ascending=False).head(10)

# Plotting
plt.figure(figsize=(12, 6))
bars = plt.bar(top_10_avg_range['Model'], top_10_avg_range['Electric Range'], color='skyblue')

# Add data labels
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom', ha='center')

plt.title('Top 10 Car Models with Highest Average Electric Range')
plt.xlabel('Car Model')
plt.ylabel('Average Electric Range (km)')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Add gridlines on the y-axis
plt.tight_layout()
plt.show()
```

```python
# Assuming df is your large DataFrame
# Extracting specific columns
columns_to_extract = ['Make', 'Model', 'Model Year', 'Electric Range']
df3 = df[columns_to_extract]

# Now df3contains only the specified columns from df
df3
```

```python
# Group by car make and calculate the average electric range
avg_range_by_make = df.groupby('Make')['Electric Range'].mean().reset_index()

# Sort the DataFrame by electric range in descending order and select the top 10 rows
top_20_avg_range = avg_range_by_make.sort_values(by='Electric Range', ascending=False).head(20)

# Plotting as a line plot
plt.figure(figsize=(12, 6))
plt.plot(top_20_avg_range['Make'], top_20_avg_range['Electric Range'], marker='o', linestyle='-')

# Add data labels
for x, y in zip(top_20_avg_range['Make'], top_20_avg_range['Electric Range']):
    plt.text(x, y, f'{y:.2f}', ha='left', va='bottom', fontsize=8)

plt.title('Top 20 Car Makes with Highest Average Electric Range')
plt.xlabel('Car Make')
plt.ylabel('Average Electric Range (km)')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add gridlines
plt.tight_layout()
plt.show()
```

```python
# Filter data for BMW vehicles
tesla_data = df[df['Make'] == 'BMW']

# Group by year and calculate the average electric range
avg_range_by_year = tesla_data.groupby('Model Year')['Electric Range'].mean().reset_index()

# Plotting
plt.figure(figsize=(8, 3))
plt.plot(avg_range_by_year['Model Year'], avg_range_by_year['Electric Range'], marker='o', linestyle='-')
plt.title('Evolution of BMW Electric Range Over Years')
plt.xlabel('Year')
plt.ylabel('Average Electric Range (km)')
plt.grid(True)  # Add gridlines
plt.tight_layout()
plt.show()
```
